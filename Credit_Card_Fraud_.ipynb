```python
import sklearn.model_selection as train_test_split
import sklearn.linear_model as LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
import pandas as pd

# Convert all columns to numeric, coercing errors to NaN
df_cleaned = df.apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values that resulted from the conversion
df_cleaned = df_cleaned.dropna()

display(df_cleaned.groupby('Class').mean())
```

```python
#Under-Sampling
#Build a sample dataset containing simple distribution of normal and fraud transaction
#Number of fraud transaction --> 492
```

```python
legit_sample = legit.sample(n=103)
```

Concatinating Two DataFrames

```python
new_data = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
import pandas as pd

# Convert all columns to numeric, coercing errors to NaN
new_data_cleaned = new_data.apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values that resulted from the conversion
new_data_cleaned = new_data_cleaned.dropna()

display(new_data_cleaned.groupby('Class').mean())
```

Splitting the data into features and targets

```python
X=new_data_cleaned.drop(columns='Class',axis=1)
Y=new_data_cleaned['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the Data into Training Data and Testing Data

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2,stratify=Y)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression.LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
model_prediction = model.predict(X_train)
```

```python
train_data_accuracy = accuracy_score(model_prediction, Y_train)
```

```python
print('Accuracy in Training Data : ',train_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy in Testing Data : ',test_data_accuracy)
```
